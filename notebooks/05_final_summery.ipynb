## Final Model Selection Summary & Configuration

After completing all experiments across Polynomial Regression, Ridge, Lasso, and Elastic Net, we consolidate the findings into a final configuration decision.

This section summarizes which model performs best, and why it is selected for the final pipeline.

---

## 1. Model Comparison Summary

| Model        | Key Strengths                          | Weaknesses                          | Validation Performance | Final Decision |
|--------------|----------------------------------------|-------------------------------------|------------------------|----------------|
| Polynomial (No Regularization) | Very low training error | Severe overfitting, unstable validation error | Very poor | ❌ Rejected |
| Ridge        | Stable, strong generalization          | Slightly higher bias               | **Best (~12.15 RMSE)** | ✅ Selected |
| Lasso        | High sparsity, feature selection       | Higher bias, lower accuracy        | Worse than Ridge       | ❌ Not selected |
| Elastic Net  | Balanced L1 + L2 behavior              | No improvement over Ridge          | Similar but weaker      | ❌ Not selected |

---

## 2. Final Hyperparameter Selection

Based on validation performance and stability:

| Component              | Selected Value |
|----------------------|----------------|
| Model                | Ridge Regression |
| Polynomial Degree   | 5 |
| Regularization (α)   | 0.01 |
| Feature Scaling      | StandardScaler (required) |
| Polynomial Features  | Included (include_bias=False) |

---

## 3. Final Model Interpretation

The selected Ridge model provides the best balance between bias and variance:

- Polynomial degree = 5 gives sufficient non-linearity
- α = 0.01 ensures minimal but effective regularization
- Model avoids both overfitting and excessive simplification

---

## 4. Key Conclusion

> The optimal model is not the most complex or the most sparse, but the one that achieves the best generalization on unseen data.

In this case, Ridge Regression with moderate regularization provides the most reliable and stable predictions for PM2.5 estimation.

---

## 5. Final Decision

✔ Selected Model → Ridge Regression  
✔ Reason → Best validation performance + stable generalization  
✔ Ready for → Production pipeline implementation
